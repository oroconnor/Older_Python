# Common Probability Distributions in Readable Python

When starting to get an understanding of the most widely used probability distributions, it can be helpful to see illustrations of the probabilty distributions created in code. This way we can understand the distributions by 'watching them happen.' This computational approach can be of particular use to those of us for whom mathmatical formulas are not our first language. 

My intention here is not to explain these probability distributions. Many sources are out there documenting the formulas, history, uses, and relationships of these common distributions, as well as showing examples. 

The purpose here is to show, in simple, readable python, how an example visualization can be created for each distribution. For each distribution, we will create a visualization of that distribution's [probability density function](https://en.wikipedia.org/wiki/Probability_density_function) and a [cumulative density function.](https://en.wikipedia.org/wiki/Cumulative_distribution_function)



```python
#first we will import a few modules that we will need.
from collections import Counter #used to create dictionaries that contain frequencies. 
import random #so we can generate different types of random numbers.
from matplotlib import pyplot as plt #we will use matplotlib to create the visualizations.
import numpy as np

#now lets write a function to make the histograms
def make_hist(sample,name):
    plt.hist(sample, edgecolor = (0,0,0))
    plt.title(name, fontsize = 15)
    plt.show
    
```

## Uniform Distribution

```python
nums = []

for i in range(100000):
    a = random.randint(1,6)
    nums.append(a)
    
histogram = Counter(nums)

plt.bar([x for x in histogram.keys()],histogram.values(),.9, edgecolor = (0,0,0))
plt.title("Uniform Distibution - Example Histogram")


plt.show
```

```python
make_hist(nums,"Uniform Distibution")
```

```python

x = [1,2,3,4,5,6]
y = [1/6,1/6,1/6,1/6,1/6,1/6]
z = []
a = 0
for i in y:
    a += i
    z.append(a)

plt.plot(x,y, "b-", label = "PMF")
plt.plot(x,z, "r-", label = "CDF")

plt.legend()
plt.title("Uniform Distibution")


plt.show
```

## Bernoulli Distribution

OK. Now we're going to create an example histogram that you might get from a problem with a bernoulli distribution. 
We're going to do a trial with a probability of success of .75. 
You can think of this as an unfair coin, where success means getting a heads.



```python

nums = [] 
for i in range(100):
    a = random.randint(1,4)
    if a > 3:
        b = 0
    else:
        b = 1
    nums.append(b)

histogram = Counter(nums)

plt.bar([x for x in histogram.keys()],histogram.values(), .1, edgecolor = (0,0,0))
plt.title("Bernoulli Distibution - Example Histogram")


plt.show
```

```python
x = [0,1,2]
y = [.25,.75,.75]
z = [.25,1,1]

plt.plot(x,y, "b-",drawstyle='steps-post', label = "PDF")
plt.plot(x,z, "r--",drawstyle='steps-post', label = "CDF")
plt.xlim(0,2)
plt.xticks((0,1))
plt.legend()
plt.title("Bernoulli Distibution")


plt.show
```

## Normal Distribution AKA Gaussian Distribution

Now were done with discrete distributions and on to continuous distributions. The first one is the Normal Distibition, or as it is otherwise known, the Gaussian Distribution. To make our histogram, were going to make use of the Normal distribution's neatest tricks, the Central Limit Theorem.

```python
population = []
sample_means = []
for x in range(100000):
    a = random.randint(0,100)
    population.append(a)

for y in range(1000):
    b = []
    for i in range(1000):
        c = random.choice(population)
        b.append(c)
    sample_mean = np.mean(b)
    sample_means.append(sample_mean)

make_hist(sample_means, "Normal Distribution")
    
```

```python
trials = []
for i in range(1000):
    head_flips = 0
    for j in range(100):
        a = random.randint(0,1)
        head_flips += a
    trials.append(head_flips)
print(trials)
make_hist(trials, "Normal Distribution")
```
